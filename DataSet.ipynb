```python
import pandas as pd
data ={'Name': ['Aman','Gauranvi','Hena','Jenna','Kumar'],
      'Age':[54,23,50,44,26],
       'Gender':['M','F','F','F','M'],
       'Marks':[90,'NaN',76,89,'NaN']
      }
df=pd.DataFrame(data)
df
```

```python
c=avg=0
for ele in df['Marks']:
    if str(ele).isnumeric():
        c+=1
        avg+=ele
        avg/=c
df=df.replace(to_replace="NaN",value=avg)
df
```

```python
df['Gender']=df['Gender'].map({'M':0,'F':2}).astype('int')
df
```

```python
df=df[df["Marks"]>=80].copy()
df
```
